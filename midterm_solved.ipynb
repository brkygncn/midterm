### 4) (10 points) Show only "Age", "Weight", "Height" and "Gender" columns of this dataset (Drop the "Education" column).

# 5) (10 points) Show dataset separately for Female and Male students.

```python
females = filteredData[filteredData["Gender"].str.contains("Female")]
females
```

```python
males = filteredData[filteredData["Gender"].str.contains("Male")]
males
```

### 6) (10 points) Print average values for Female and Male students separately.

```python
print("Average age of female students: " + str(females.mean(axis=0)[0]))
print("Average age of male students: " + str(males.mean(axis=0)[0]))

print("Average weight of female students: " + str(females.mean(axis=0)[1]))
print("Average weight of male students: " + str(males.mean(axis=0)[1]))

print("Average height of female students: " + str(females.mean(axis=0)[2]))
print("Average height of male students: " + str(males.mean(axis=0)[2]))
```

### 5) (20 points) Draw the scatter plot of the dataset using attributes "Weight" and "Height" according to class label "Gender". Use different colors for "Gender" on the plot.

```python
import matplotlib.pyplot as plt
groups=table.groupby("Gender")
for name,group in groups:
    plt.plot(group["Weight"], group["Height"], marker="o", linestyle="", label=name)
plt.legend()
plt.title("Attribute Values of Students")
```
